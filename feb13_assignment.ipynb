Q4

LookupError class exceptions are raised when an index or a key is not found for a sequence or dictinary respectively.



```python
# 1 IndexError
try :
    l = [3,4,56,6,7]
    print(l[20])
except Exception as e:
    print(e,e.__class__)
    
# IndexError gets when we access an item from list which is out of range
```

```python
# 2 KeyError
try:
    d = {"key1":"Hello","Key2":"World"}
    print(d["key3"])
except Exception as e:
    print(e,e.__class__)

# KeyError gets when we access the key in the dictionary which doesn't exist.
```
